<h1><center>Welcome to SHINYCAT alpha testing!</center></h1>

Thank you for your willingness to participate in the SHINYCAT alpha testing.  SHINYCAT is an interactive application that delivers various forms of the Philadelphia Naming Test (PNT) and outputs item level test results along with an Item Response Theory (IRT) naming ability estimate for each participant. As intended in its namesake, SHINYCAT also provides computer adaptive testing (CAT) versions of the PNT. Availble testing instruments include:
- The original PNT as published by [Roach et al 1996](http://aphasiology.pitt.edu/215/1/24-09.pdf)
- A 30 item computer adapative test as published by [Fergadiotis et al, 2018](https://doi.org/10.1044/2018_JSLHR-L-18-0344)
- A Variable length PNT intended to be delivered after the 30 item PNT as published by [Hula et al, 2019](https://doi.org/10.1044/2019_JSLHR-L-19-0213)

In addition, there is an option to run longer computer adaptive tests at a length of 60 and 100 items.

**The intended purpose** for this testing is not to inform the volunteers of the psychometric properties, but to challenge volunteers to navigate through the application as if they were testing the naming ability of individuals with aphasia. Future documents of the psychometric frame of the assessments provided in SHINYCAT will be released upon further testing of the application.

We ask the volunteer to provide any and all feedback of the application.  At this time, **reporting of problems, bugs, typos, and any type of confusions incountered by the viewer** be documented and shared with us.  

Please include the step number with all comments to help us identify where problems may be occuring.
